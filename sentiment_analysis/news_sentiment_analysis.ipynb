```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
import pandas as pd
```

```python
#Datasets
real_df = pd.read_csv("https://raw.githubusercontent.com/joshndala/fake-news-detection/main/news_dataset/True.csv")
fake_df = pd.read_csv("https://raw.githubusercontent.com/joshndala/fake-news-detection/main/news_dataset/Fake.csv")

real_df = real_df.sample(frac=0.15, random_state=2024)
fake_df = fake_df.sample(frac=0.15, random_state=2024)
```

```python
#Load model and tokenizer
model_name = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
```

```python
#Load pipeline
sentiment_pipeline = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer)
```

```python
#Function for sentiment analysis
def analyze_sentiment(text, max_length=512):
  #Truncate the text to the max length
  truncated_text = text[:max_length]
  result = sentiment_pipeline(truncated_text)[0]
  return result['label'], result['score']
```

```python
#Apply sentiment analysis function to both datasets
real_df['Sentiment'], real_df['Sentiment_Score'] = zip(*real_df['text'].apply(analyze_sentiment))
fake_df['Sentiment'], fake_df['Sentiment_Score'] = zip(*fake_df['text'].apply(analyze_sentiment))
```

```python
# Average sentiment score by News Type (Real or Fake)
average_sentiment_score_real = real_df['Sentiment_Score'].mean()
average_sentiment_score_fake = fake_df['Sentiment_Score'].mean()

print("Average Sentiment Score for Real News:", average_sentiment_score_real)
print("Average Sentiment Score for Fake News:", average_sentiment_score_fake)
```
